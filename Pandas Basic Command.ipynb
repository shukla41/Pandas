```python
## Iloc function 
## Making a column as index and use loc method  to fetch rows for particular value of the index.
data=pd.read_csv(r'C:\Users\shumondal\Downloads\Video-Lecture-NBs\Video-Lecture-NBs\Video_Lectures\summer.csv', index_col='Year')
data.loc["2012"]
```
